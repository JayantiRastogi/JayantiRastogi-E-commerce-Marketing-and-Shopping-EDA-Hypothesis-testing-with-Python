##Univariate Analysis

```python
# Univariate analysis for checking outliers

outlier_columns_check = df[["Administrative_Duration", "Informational_Duration", "ProductRelated_Duration", "BounceRates", "ExitRates", "PageValues"]]

plt.figure(figsize = (18, 6))
sns.boxplot(outlier_columns_check)
plt.show()
```

```python
# Cap outliers using IQR Method
outliers_columns_cap = ["Administrative_Duration", "Informational_Duration", "ProductRelated_Duration", "BounceRates", "ExitRates", "PageValues"]

Q1 = df[outliers_columns_cap].quantile(0.25)
Q3 = df[outliers_columns_cap].quantile(0.75)

for column in df[outliers_columns_cap]:
  df[column] = np.where(df[column] < Q1[column], Q1[column], df[column])
  df[column] = np.where(df[column] > Q3[column], Q3[column], df[column])
```

## Summarize page views, durations, and bounce/exit rates for each page category.


```python
# Page Catgorization

page_category_summary = df.groupby(["Administrative", "Informational", "ProductRelated"]).agg({
    "Administrative_Duration": ["mean", "sum"],
    "Informational_Duration": ["mean", "sum"],
    "ProductRelated_Duration": ["mean", "sum"],
    "BounceRates": ["mean"],
    "ExitRates": ["mean"],
    "PageValues": ["mean"]
}).reset_index()

page_category_summary.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in page_category_summary.columns]
page_category_summary
```

```python
# Melt the summary DataFrame for easier plotting

page_category_summary.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in page_category_summary.columns]

melted_df = page_category_summary.melt(id_vars=['Administrative_', 'Informational_', 'ProductRelated_'],
                                        value_vars=['Administrative_Duration_mean', 'Administrative_Duration_sum',
                                                    'Informational_Duration_mean', 'Informational_Duration_sum',
                                                    'ProductRelated_Duration_mean', 'ProductRelated_Duration_sum',
                                                    'BounceRates_mean', 'ExitRates_mean', 'PageValues_mean'],
                                        var_name='Metric', value_name='Value').reset_index(drop=True)

melted_df
```

```python
# Create subplots
fig, axes = plt.subplots(3, 4, figsize=(18, 10))

# Plot each metric
sns.barplot(x='Administrative_', y='Administrative_Duration_mean', ax=axes[0, 0], data=page_category_summary)
axes[0, 0].set_title('Admin Duration')

sns.barplot(x='Informational_', y='Informational_Duration_mean', ax=axes[0, 1], data=page_category_summary)
axes[0, 1].set_title('Info Duration')

sns.barplot(x='ProductRelated_', y='ProductRelated_Duration_mean', ax=axes[0, 2], data=page_category_summary)
axes[0, 2].set_title('Product Duration')

sns.barplot(x='Administrative_', y='BounceRates_mean', ax=axes[0, 3], data=page_category_summary)
axes[0,3].set_title('Bounce Rates')

sns.barplot(x='Administrative_', y='ExitRates_mean', ax=axes[1, 0], data=page_category_summary)
axes[1, 0].set_title('Exit Rates')

sns.barplot(x='Administrative_', y='PageValues_mean', ax=axes[1, 1], data=page_category_summary)
axes[1, 1].set_title('Page Values')

sns.barplot(x='Informational_', y='BounceRates_mean', ax=axes[1, 2], data=page_category_summary)
axes[1, 2].set_title('Page Values')

sns.barplot(x='Informational_', y='ExitRates_mean', ax=axes[1, 3], data=page_category_summary)
axes[1, 3].set_title('Page Values')

sns.barplot(x='Informational_', y='PageValues_mean', ax=axes[2, 0], data=page_category_summary)
axes[2, 0].set_title('Page Values')

sns.barplot(x='ProductRelated_', y='PageValues_mean', ax=axes[2, 1], data=page_category_summary)
axes[2, 1].set_title('Page Values')

sns.barplot(x='ProductRelated_', y='BounceRates_mean', ax=axes[2, 2], data=page_category_summary)
axes[2, 2].set_title('Page Values')

sns.barplot(x='ProductRelated_', y='ExitRates_mean', ax=axes[2, 3], data=page_category_summary)
axes[2, 3].set_title('Page Values')

# Adjust layout
plt.tight_layout()
plt.show()
```

## Group users based on VisitorType, OperatingSystems, and Region to identify potential differences in behavior and conversion rates.

```python
# Group users based on VisitorType, OperatingSystems, and Region to identify potential differences in behavior and conversion rates.

df['VisitorType'] = df['VisitorType'].astype('category')

# Grouping users based on VisitorType, OperatingSystems, and Region
grouped_df = df.groupby(['VisitorType', 'OperatingSystems', 'Region']).agg({
    'PageValues': ['mean', 'sum'],
    'Revenue': 'mean',  # Calculating the conversion rate as the mean of Revenue (1 for conversion, 0 for no conversion)
    'Administrative_Duration': 'mean',
    'Informational_Duration': 'mean',
    'ProductRelated_Duration': 'mean'
}).reset_index()

# Renaming the columns for clarity
grouped_df.columns = ['VisitorType', 'OperatingSystems', 'Region',
                      'AvgPageValues', 'TotalPageValues', 'ConversionRate',
                      'AvgAdministrativeDuration', 'AvgInformationalDuration', 'AvgProductRelatedDuration']

# Display the resulting grouped DataFrame
grouped_df
```

```python
fig , axes = plt.subplots(1, 3, figsize = (18,5))

axes = axes.flatten()

sns.violinplot(x = grouped_df["VisitorType"], y = grouped_df["ConversionRate"], ax = axes[0])
axes[0].set_title('Conversion Rate by Visitor Type')

sns.barplot(x = grouped_df["Region"], y = grouped_df["ConversionRate"], ax = axes[1])
axes[1].set_title('Conversion Rate by Region')

sns.lineplot(x = grouped_df["OperatingSystems"], y = grouped_df["ConversionRate"], ax = axes[2])
axes[2].set_title('Conversion Rate by Operating Systems')

plt.show()
```

## Segment users based on TrafficType and analyze their engagement patterns and purchase probability.


```python
traffic_type_summary = df.groupby('TrafficType').agg({
    'PageValues': ['mean', 'sum'],
    'SessionLength': ['mean'],
    'Revenue': 'mean'
}).reset_index()
traffic_type_summary.columns = ['TrafficType', 'AvgPageValues', 'TotalPageValues', 'AvgSessionLength', 'ConversionRate']

traffic_type_summary
```

```python
fig , axes = plt.subplots(1, 2, figsize = (18,5))

axes = axes.flatten()

sns.lineplot(x = traffic_type_summary["TrafficType"], y = traffic_type_summary["ConversionRate"], ax = axes[0])
axes[0].set_title('Conversion Rate by Traffic Type')

sns.lineplot(x = traffic_type_summary["TrafficType"], y = traffic_type_summary["AvgSessionLength"], ax = axes[1])
axes[1].set_title('Avg Session Length by Traffic Type')

plt.show()
```
